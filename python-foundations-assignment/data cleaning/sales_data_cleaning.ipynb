```python
import pandas as pd
import numpy as np
data = {
'Order_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
'Customer_Name': ['Amara', 'Bola', None, 'Diana', 'Efua', 'Fatima', 'Grace',
None, 'Ifeoma', 'Joyce'],
'Country': ['Ghana', 'Nigeria', 'Kenya', None, 'Ghana', 'Cameroon', None,
'Kenya', 'Nigeria', 'Ghana'],
'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', None, 'Phone', 'Tablet',
'Laptop', None, 'Phone'],
'Quantity': [2, 1, None, 3, 2, None, 1, 4, 2, None],
'Price': [1500, 800, 600, 1500, None, 800, 600, 1500, None, 800],
'Delivery_Days': [3, None, 7, 5, 3, None, 6, None, 4, 5]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.info()
df.describe()
df.isnull().sum()



#There are 7 columns in total (Order_ID, Customer_Name, Country, Product, Quantity, Price, and Delivery_Days)
#Customer_Name, Country, Product, and  Price have two null values each.
#Quantity and Delivery_Days have three null values each 
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
print(df)

# I used "Unknown" to label missing customer names because it 
# prevents pipeline errors while keeping the row intact.
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price'] = df['Price'].fillna(df['Price'].median())

# I used median for both coulmns.
# Using median for null numerical values is best practice as it is robust and handles fututre datasets better.
print(df)
```

```python
print(df['Delivery_Days'].describe())

df['Delivery_Days'] = df['Delivery_Days'].fillna(df['Delivery_Days'].median())

# With values between 3 and 7 days, there are no outliers. 
# I chose the median because delivery days are whole numbers,
# and the mean might produce an decimal, i.e (4.5 days).
```

```python
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
df['Product'] = df['Product'].fillna(df['Product'].mode()[0])

# I filled the missing values to avoid dropping rows and losing revenue data. 
# I used the mode because it is the standard method for category columns and preserves the dataset's natural distribution.
```
