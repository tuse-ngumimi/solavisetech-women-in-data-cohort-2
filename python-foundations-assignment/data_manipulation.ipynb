```python
import pandas as pd

data = {
    'Order_ID': ['ORD001','ORD002','ORD003','ORD004','ORD005',
    'ORD006','ORD007','ORD008','ORD009','ORD010'],
    'Product': ['Laptop','Rice Bag','Headphones','Notebook','Blender',
    'Rice Bag','Laptop','Notebook','Blender','Headphones'],
    'Category': ['Electronics','Food','Electronics','Stationery','Appliances',
    'Food','Electronics','Stationery','Appliances','Electronics'],
    'Region': ['North','South','East','West','North',
    'East','South','North','West','North'],
    'Units_Sold': [5, 30, 12, 50, 8, 20, 3, 80, 6, 15],
    'Unit_Price': [850.00, 12.50, 45.00, 3.20, 65.00,
    12.50, 850.00, 3.20, 65.00, 45.00],
    'Month': ['January','January','February','February','March',
    'March','April','April','May','May']
}
df = pd.DataFrame(data)
print(df)
```

```python
print(f"\n The Column below includes the Product and Region columns. \n")

print(df[['Product', 'Region']])
print("\n")
print(f"\n The Column below includes the Order_ID, Category and Units_Sold columns. \n")

print(df[['Order_ID', 'Category','Units_Sold']])
print("\n")

print(f"This is the Unit_Price column using a single bracket: \n", df['Unit_Price'])
#This output is a one dimensional labelled array 
print("\n")

print(f"This is the Unit_Price column using double brackets: \n", df[['Unit_Price']])
#This output is a 2D tabular structure that includes rows and colunms 
```

```python
print("Below are the rows where the region is North: \n", df.loc[df['Region'] == 'North'])
print("\n")

print("Below are the rows where the units sold are more than 10: \n", df.loc[df['Units_Sold'] > 10])
print("\n")

print("Below are the rows where the categories are Electronics and Units_Sold are greater than 5: \n", df.loc[(df['Category'] == 'Electronics') & (df['Units_Sold'] > 5)])
print("\n")

print("Below are the rows where the Region is South or Region is West: \n", df.loc[(df['Region'] == 'South') | (df['Region'] == 'West')])
```

```python
total_units_sold_per_region = df.groupby('Region')['Units_Sold'].sum().reset_index()
print("Total Units Sold per Region:\n", total_units_sold_per_region)

print("\n")
avg_unit_price_per_category = df.groupby('Category')['Unit_Price'].mean().reset_index()
print("Average Unit Price per Category:\n", avg_unit_price_per_category)

print("\n")
category_summary = df.groupby('Category').agg(
    total_units_sold = ('Units_Sold', 'sum'),
    average_unit_price = ('Unit_Price', 'mean')
    ).reset_index()
print("Summary for Total Units Sold and Average Unit Price):\n", category_summary)

print("\n")
total_units_sold_per_month = df.groupby('Month')['Units_Sold'].sum().reset_index()
print("Total Units Sold per Month:\n", total_units_sold_per_month)

print("\n")
total_units_per_region = df.groupby('Region')['Units_Sold'].sum()
highest_units_sold_region = total_units_per_region.idxmax()
print(f"The region with the highest total Units_Sold is: {highest_units_sold_region}")

print("\n")
total_units_sold_per_month = df.groupby('Month')['Units_Sold'].sum().reset_index()
print("Total Units Sold per Month:\n", total_units_sold_per_month)
```

```python
product_avg_price = df.groupby('Product')['Unit_Price'].mean()
lowest_avg_price_product = product_avg_price.idxmin()
print(f"The product with the lowest average Unit_Price is: {lowest_avg_price_product}")

print("\n")
month_total_units = df.groupby('Month')['Units_Sold'].sum()
highest_units_sold_month = month_total_units.idxmax()
print(f"The month with the highest total Units_Sold is: {highest_units_sold_month}")

print("\n")
filtered_orders = df[(df['Category'] == 'Electronics') & (df['Region'] == 'North')]
display_columns = filtered_orders[['Product', 'Units_Sold', 'Unit_Price']]
print("Electronics orders from the North region:\n", display_columns)
```
