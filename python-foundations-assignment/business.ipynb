```python
import pandas as pd

data = {
    "Region":   ["North","South","East","West","North","East","South","West"],
    "Category": ["Electronics","Clothing","Electronics","Clothing",
                 "Food","Food","Electronics","Food"],
    "Month":    ["Jan","Jan","Feb","Feb","Mar","Mar","Apr","Apr"],
    "Sales":    [1200, 800, 1500, 600, 900, 1100, 1300, 750],
    "Units":    [12, 20, 15, 18, 25, 22, 13, 30]
}

df = pd.DataFrame(data)
print(df)
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
highest_revenue_region = region_sales.idxmax()
max_revenue = region_sales.max()
print(f"The region with highest total revenue is {highest_revenue_region} - Sales: ${max_revenue}")

category_units = df.groupby('Category')['Units'].sum()
highest_units_category = category_units.idxmax()
max_units = category_units.max()
print(f"The product category with most units sold is {highest_units_category} - Units: {max_units}")

month_sales = df.groupby('Month')['Sales'].sum()
highest_sales_month = month_sales.idxmax()
max_month_sales = month_sales.max()
print(f"The month with highest total sales is {highest_sales_month} - Sales: ${max_month_sales}")

electronics_sales = df[df['Category'] == 'Electronics']['Sales']
avg_electronics_sales = electronics_sales.mean()
print(f"The average sale value for electronics category is  ${avg_electronics_sales:.2f}")

region_avg_sales = df.groupby('Region')['Sales'].mean()
lowest_avg_sales_region = region_avg_sales.idxmin()
min_avg_sales = region_avg_sales.min()
print(f"The region with lowest average sales is {lowest_avg_sales_region} - Average Sales: ${min_avg_sales:.2f})")
```
