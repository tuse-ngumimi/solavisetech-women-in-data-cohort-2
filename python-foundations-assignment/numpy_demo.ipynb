```python
import numpy as np
```

```python
### Homework Assignment

expenses = np.array([1500, 2000, 1800, 2200, 1700, 2500])

average_monthly_exp = expenses.mean()

max_monthly_exp = expenses.max()

min_monthly_exp = expenses.min()

total_monthly_exp = expenses.sum()

first_three_sales = expenses[0:3]

print(f"The Average, Highest, Lowest and Total Sales: {average_monthly_exp}, {max_monthly_exp}, {min_monthly_exp}, and {total_monthly_exp} respectively")

print(f"The first three monthly sales include: {first_three_sales}")
```

```python
sales = np.array([500, 700, 900, 1200, 1500])

sales_avg = np.average(sales)

sales_max = np.max(sales)

sales_min = np.min(sales)

print(sales_avg)

print(sales_max)

print(sales_min)
```

```python
monthly_sales = np.array([500, 700, 900, 1200, 1500, 1800])

average_sales = monthly_sales.mean()

highest_sales = monthly_sales.max()

lowest_sales = monthly_sales.min()

total_sales = monthly_sales.sum()


print(f"The Average, Highest, Lowest and Total Sales: {average_sales}, {highest_sales}, {lowest_sales}, and {total_sales} respectively")

first_five_sales = monthly_sales[0:5]
print(f"The first five monthly sales are {first_five_sales}")
```
